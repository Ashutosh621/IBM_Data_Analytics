
 ### Importing necessory library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Dataset

```python
df = pd.read_csv("Superstoredata.csv")
df.head()
```

### Describing Info

```python
df.info()
```

```python
df.isnull().sum()
```

# Exploratory Data Analysis - EDA

```python
df.dtypes
```

### Checking for the Duplicate values

```python
df.duplicated().sum()
```

### Droping the duplicate values

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

### Droping the Column not needed

```python
df=df.drop('Postal Code', axis=1)
```

### Plotting the number of Orders for each Segment

```python
plt.figure(figsize=(20, 5))
plt.title("Number of orders for each Segment")
sns.countplot(data=df, x='Segment')
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

### Plotting the Segment and their Sales

```python
plt.figure(figsize=(20,5))
plt.title("Segment and their Sales")
sns.barplot(x = df['Segment'] , y = df["Sales"])
```

### Plotting the Segment and the Profit

```python
plt.figure(figsize=(20,5))
plt.title("Segment and their profit")
sns.barplot(x = df['Segment'] , y = df["Profit"])
```

### Plotting the number of order for each ship mode

```python
plt.figure(figsize=(20, 5))
plt.title("Number of orders for each Ship Mode")
sns.countplot(data=df, x='Ship Mode')
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.show()
```

### Plotting Ship mode and their Sale

```python
plt.figure(figsize=(20,5))
plt.title("Ship mode and their sales")
sns.barplot(x = df['Ship Mode'] , y = df["Sales"])
```

### Plotting ship mode and their profit

```python
plt.figure(figsize=(20,5))
plt.title("Ship mode and their profit")
sns.barplot(x = df['Ship Mode'] , y = df["Profit"])
```

### Plotting the heat map for the Correlation

```python
plt.figure(figsize=(20,5))
sns.heatmap(df.corr(), cmap = 'Blues', annot = True)
plt.show()
```

### Plotting the Scatter Plot for Sale v/s Profit

```python
plt.scatter(df['Sales'],df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales v/s Profit')
plt.show()
```

### Plotting the count plot for Number of order by region

```python
plt.figure(figsize=(20,5))
plt.title("Number of orders by Region")
sns.countplot(data=df, x='Region')
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

### Bar Plot showing the Region wise Sale v/s Profit

```python
pd.DataFrame(df.groupby('Region').sum()[['Sales','Profit']]).plot(kind='bar')
```

### Bar plot showing number of order by each state

```python
plt.figure(figsize=(20,5))
plt.title("Number of orders by each state")
sns.countplot(data=df, x='State')
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

### Bar plot showing State wise Profit v/s Sales

```python
pd.DataFrame(df.groupby('State').sum()[['Sales','Profit']]).plot(kind='bar',figsize=(20,5))
```

### Bar plot showing Top 5 city by Sales

```python
city_Sale = pd.DataFrame(df.groupby('State').sum()['Sales']).sort_values(by='Sales', ascending=False)
city_Sale.reset_index(inplace=True)
top_5_cities_Sales =city_Sale.head()
plt.bar(top_5_cities_Sales['State'], top_5_cities_Sales['Sales'], align='center')
plt.xlabel("State")
plt.ylabel("Sales")
plt.title("Top 5 Cities by Sales")
plt.xticks(rotation=90)
plt.show()
top_5_cities_Sales
```

### Bar plot showing City with minimum amount of sale

```python
city_Sale = pd.DataFrame(df.groupby('State').sum()['Sales']).sort_values(by='Sales', ascending=True)
city_Sale.reset_index(inplace=True)
top_5_cities_Sales =city_Sale.head()
plt.bar(top_5_cities_Sales['State'], top_5_cities_Sales['Sales'], align='center')
plt.xlabel("City")
plt.ylabel("Sales")
plt.title("Minimum amount of Sales")
plt.xticks(rotation=90)
plt.show()
top_5_cities_Sales
```

### Bar plot showing Top 5 city by Profit

```python
city_profit = pd.DataFrame(df.groupby('State').sum()['Profit']).sort_values(by='Profit', ascending=False)
city_profit.reset_index(inplace=True)
top_5_cities_profit =city_profit.head()
plt.bar(top_5_cities_profit['State'], top_5_cities_profit['Profit'], align='center')
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("Top 5 Cities by Profit")
plt.xticks(rotation=90)
plt.show()
top_5_cities_profit
```

### Bar plot showing City with minimum amount of Profit

```python
city_loss = pd.DataFrame(df.groupby('State').sum()['Profit']).sort_values(by='Profit', ascending=True)
city_loss.reset_index(inplace=True)
top_5_cities_loss =city_loss.head()
plt.bar(top_5_cities_loss['State'], top_5_cities_loss['Profit'], align='center')
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("City by minimum Profit")
plt.xticks(rotation=90)
plt.show()
top_5_cities_loss
```

### Pie plot showing the Proportion of total sales and profit

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
labels = ['Total Sales', 'Total Profit']
sizes = [total_sales, total_profit]
colors = ['purple', 'green']
explode = (0, 0.1)
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Proportion of Total Sales and Total Profit')
plt.show()
```

### Bar plot is used to show most sold category

```python
category_sales = df.groupby('Category')['Sales'].sum()
most_sold_category = category_sales.idxmax()
plt.figure(figsize=(10, 6))
plt.bar(category_sales.index, category_sales.values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales for Each Category')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.bar(most_sold_category, category_sales.loc[most_sold_category], color='red')
plt.show()
```
